##Functions

```python
def extractSchoolContact(soup_details):
  address, phone, fax, website, chairman, principal = ["" for i in range(6)]
  try:
    for p in soup_details.select("div.contact > p"):
      tokens = p.get_text().split("\n")
      for token in tokens:
        if "地址" in token:
          address = token.replace("地址：","").strip()
          
        if "電話" in token or "傳真" in token:
          phone = token.split("\xa0")[0].replace("電話：","").strip()
          fax = token.split("\xa0")[1].replace("傳真：","").strip()
        
        if "網址" in token:
          website = token.replace("網址：","").strip()

        if "校監" in token and "校長" in token:
          chairman = token.split("校長")[0].replace("校監：","").strip()
          principal = token.split("校長")[1].replace("：","").strip()
  except:
    pass
        
  # print(address, phone, website, fax, chairman, principal, sep=" | ")
  return address, phone, website, fax, chairman, principal

def extractTableNumber(soup_details):
  counter = 1
  tableNumber = soup_details.select("table.education-state")[0]
  dctSchooltypeMap = {"上午班":"AM", "下午班":"PM", "全日班":"WD"}
  numberCounter = 0
  data = {}
  for t in tableNumber.select("tbody > tr"):
    lstNumber = []
    try:
      schooltype = dctSchooltypeMap[t.find("th").get_text().strip()]
      tdLst = t.find_all("td")
    except:
      schooltype = dctSchooltypeMap[t.find("td").get_text().strip()]
      tdLst = t.find_all("td")[1:]

    for td in tdLst:
      lstNumber.append(td.get_text().replace(",","").strip())
    data[schooltype] = lstNumber
    numberCounter += 1
  return data

def extractTablePrice(soup_details):
  counter = 1
  tablePrice = soup_details.select("table.education-state")[1]
  dctSchooltypeMap = {"上午班":"AM", "下午班":"PM", "全日班":"WD"}
  priceCounter = 0
  data = {}
  for t in tablePrice.select("tbody > tr"):
    lstPrice = []
    schooltype = dctSchooltypeMap[t.find("td").get_text().strip()]
    for td in t.find_all("td")[1:]:
      # print(td.get_text().replace("$","").strip())
      lstPrice.append(removeParanthesesContent(td.get_text().replace("$","").replace(",","").replace("免費","0").strip()))
    data[schooltype] = lstPrice
    priceCounter += 1
  return data

def removeParanthesesContent(string):
  cleanrule = re.compile("\(.*?\)")
  string = re.sub(cleanrule,"",string)
  return string
```

##Main Scraper

```python
entries = []
rowcounter = 1
total = len(records)
print("Scraping ... ")
for row in records:
  link = base_url + row[0][row[0].find('"') + 1:row[0].find('">')]
  capacity = row[5]

  details = requests.get(link)
  soup_details = BeautifulSoup(details.content, 'html.parser')

  #attributes
  district = soup_details.select("div#page-heading > h1")[0].get_text().strip()
  chinese_name = soup_details.select("div.desc > h2 > span.chinese-name")[0].get_text().strip()
  english_name = soup_details.select("div.desc > h2 > span.english-name")[0].get_text().strip()

  print("- (",rowcounter,"/",total,") chinese_name: ",chinese_name, "| english_name: ", english_name, " ... ", end=" ")

  description, schooltype_scheme, facilities, history_description = "", "", "", ""
  found_year, staff, staff_student_ratio = "", "", ""

  try:
    description = soup_details.select("div.desc > p")[0].get_text().strip()
    schooltype_scheme = soup_details.find("h3", text="學校類別").parent.nextSibling.nextSibling.find("h4").get_text().strip()
    facilities = soup_details.find("h3", text="學校設施").parent.nextSibling.nextSibling.nextSibling.nextSibling.find("p").get_text().strip()
    history_description = soup_details.find("h3", text="創校歷史").parent.nextSibling.nextSibling.nextSibling.nextSibling.find("p").get_text().strip()
  except:
    pass

  try:
    found_year = soup_details.find("h3", text="創校歷史").parent.nextSibling.nextSibling.find("h4").get_text().strip().split(" ")[0]
  except:
    pass
  
  try:
    staff = [dig for dig in soup_details.find("h3", text="教學情況").parent.nextSibling.nextSibling.get_text().strip().split(" ") if dig.isdigit()][0]
    staff_student_ratio = [dig for dig in soup_details.find("h3", text="教學情況").parent.nextSibling.nextSibling.get_text().strip().split(" ") if dig.isdigit()][-1]
  except:
    pass

  address, phone, website, fax, chairman, principal = extractSchoolContact(soup_details)

  # Number_Age34_AM, Number_Age45_AM, Number_Age56_AM, Number_Total_AM = "", "", "", ""
  # Number_Age34_PM, Number_Age45_PM, Number_Age56_PM, Number_Total_PM = "", "", "", ""
  # Number_Age34_WD, Number_Age45_WD, Number_Age56_WD, Number_Total_WD = "", "", "", ""

  # try:
  tableNumber = extractTableNumber(soup_details)
  Number_Age34_AM, Number_Age45_AM, Number_Age56_AM, Number_Total_AM = tableNumber['AM']
  Number_Age34_PM, Number_Age45_PM, Number_Age56_PM, Number_Total_PM = tableNumber['PM']
  Number_Age34_WD, Number_Age45_WD, Number_Age56_WD, Number_Total_WD = tableNumber['WD']
  # except:
  #   pass
  
  Price_Age34_AM, Price_Age45_AM, Price_Age56_AM = "", "", ""
  Price_Age34_PM, Price_Age45_PM, Price_Age56_PM = "", "", ""
  Price_Age34_WD, Price_Age45_WD, Price_Age56_WD = "", "", ""

  try:
    tablePrice = extractTablePrice(soup_details)
    Price_Age34_AM, Price_Age45_AM, Price_Age56_AM = tablePrice['AM']
    Price_Age34_PM, Price_Age45_PM, Price_Age56_PM = tablePrice['PM']
    Price_Age34_WD, Price_Age45_WD, Price_Age56_WD = tablePrice['WD']
  except:
    pass

  result = {
      "district":district,
      "chinese_name": chinese_name,
      "english_name": english_name,
      "description": description,
      "schooltype_scheme": schooltype_scheme,
      "facilities": facilities,
      "found_year": found_year,
      "history_description": history_description,
      "capacity": capacity,
      "staff": staff,
      "staff_student_ratio": staff_student_ratio,
      "address": address,
      "phone": phone,
      "website": website,
      "fax": fax,
      "chairman": chairman,
      "principal": principal,
      "Number_Age34_AM": Number_Age34_AM,
      "Number_Age45_AM": Number_Age45_AM,
      "Number_Age56_AM": Number_Age56_AM,
      "Number_Total_AM": Number_Total_AM,
      "Number_Age34_PM": Number_Age34_PM,
      "Number_Age34_PM": Number_Age34_PM,
      "Number_Age45_PM": Number_Age45_PM,
      "Number_Age56_PM": Number_Age56_PM,
      "Number_Total_PM": Number_Total_PM,
      "Number_Age34_WD": Number_Age34_WD,
      "Number_Age45_WD": Number_Age45_WD,
      "Number_Age56_WD": Number_Age56_WD,
      "Number_Total_WD": Number_Total_WD,
      "Price_Age34_AM": Price_Age34_AM,
      "Price_Age45_AM": Price_Age45_AM,
      "Price_Age56_AM": Price_Age56_AM,
      "Price_Age34_PM": Price_Age34_PM,
      "Price_Age45_PM": Price_Age45_PM,
      "Price_Age56_PM": Price_Age56_PM,
      "Price_Age34_WD": Price_Age34_WD,
      "Price_Age45_WD": Price_Age45_WD,
      "Price_Age56_WD": Price_Age56_WD
  }
  entries.append(result)
  print("done")
  rowcounter += 1
print("Complete.")
```
